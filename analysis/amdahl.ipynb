```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
root_directory = os.path.dirname("../")
logs_directory = os.path.join(root_directory, "logs")
current_dir = os.path.dirname(".")
plots_directory = os.path.join(current_dir, "plots")
```

```python
fig, ax = plt.subplots(figsize=(8, 6), dpi=300)
x = np.arange(0, 17)
p_1 = 0.38
y_p1 = 1/((1-p_1) + p_1/x)

p_2 = 0.865
y_p2 = 1/((1-p_2) + p_2/x)
ax.plot(x, y_p1, zorder=3, linestyle="-.", label=f"Theoretical, $p={p_1:.2f}$")
ax.plot(x, y_p2, zorder=3, linestyle="--", label=f"Theoretical, $p={p_2:.2f}$")
#ax.scatter([2, 4], [1.66, 2.74], zorder=3, linestyle="--", label=f"mcMPQUIC", color='red', marker='*')
ax.plot(x, x, zorder=3, linestyle=":", label="Ideal")
#ax.plot(rates, exponential_theory, label="Exponential", zorder=3, linestyle="-.", marker="*")
ax.set_xlabel('CPU cores')
ax.set_ylabel('Speedup')
ax.set_title(f'Theoretical Speedup, Amdahl\'s Law')
#ax.set_title(f'Observed Speedup, Amdahl\'s Law')
ax.set_xticks(x,x )
ax.set_xlim(left=1)
ax.set_ylim(bottom=0)
ax.legend(title='Parallelism percentage')
ax.grid(which='major', zorder=0, linestyle="dashed")  # Show the major grid behind the plot
plt.savefig(os.path.join(plots_directory, "profiling_mcmpquic_amdahl.pdf"), format='pdf', bbox_inches='tight')
plt.savefig(os.path.join(plots_directory, "profiling_mcmpquic_amdahl.svg"), format='svg', bbox_inches='tight')
```
