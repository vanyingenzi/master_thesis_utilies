```python
import numpy as np
import matplotlib.pyplot as plt
import os
import json
from typing import List, Dict, Any, Tuple
from datetime import datetime  
import re
from collections import defaultdict
```

```python
DPI=300
```

```python
root_directory = os.path.dirname("../")
logs_directory = os.path.join(root_directory, "logs")
expirements_logs = os.path.join(logs_directory, "logs_2024-05-23T10:47:38")
current_dir = os.path.dirname(".")
plots_directory = os.path.join(current_dir, "plots")
testcase = "throughput" # "throughput" or "goodput"
```

```python
def get_test_start_end_time(time_json_file) -> Tuple[int, int]:
    with open(time_json_file) as f:
        data = json.load(f)
        return data["start"], data["end"]
```

```python
def is_valid_timestamp(timestamp):
    pattern = r"^[0-2][0-9]:[0-5][0-9]:[0-5][0-9]$"
    return bool(re.match(pattern, timestamp))

def extract_ifstat_data_file(filename: str, start_time: int, end_time: int, utah_time) -> List[float]:
    with open(filename, "r") as f:
        lines = f.readlines()
    lines = [line.strip() for line in lines if len(line.strip()) != 0]
    start_datetime = datetime.fromtimestamp(int(start_time / 1e9))
    end_datetime = datetime.fromtimestamp(int(end_time / 1e9))
    data_per_second = []
    for line in lines:
        timestamp = line.split()[0]
        if not is_valid_timestamp(timestamp):
            continue
        timestamp_datetime = datetime.strptime(timestamp, '%H:%M:%S')
        timestamp_datetime = timestamp_datetime.replace(year=start_datetime.year, month=start_datetime.month, day=start_datetime.day)
        if utah_time:
            timestamp_datetime = timestamp_datetime.replace(hour=timestamp_datetime.hour + 7)
        if not (start_datetime <= timestamp_datetime and timestamp_datetime <= end_datetime):
            continue
        data = sum([float(measurement) for measurement in line.split()[1:]])
        data_per_second.append(data)
    sum_data = sum(data_per_second)
    #print(f"Total data: {sum_data / (1024 * 1024)} GB")
    return np.array(data_per_second) * 8 / 1024 # Convert to Mbps

def extract_ifstat_data_per_role(role_logs: str, start, end, utah_time: bool):
    mpstat_files = [f for f in os.listdir(role_logs) if f.startswith("ifstat_monitor")]
    if len(mpstat_files) == 0:
        raise Exception("No mpstat file found")
    elif len(mpstat_files) > 1:
        raise Exception("More than one mpstat file found")
    return extract_ifstat_data_file(os.path.join(role_logs, mpstat_files[0]), start, end, utah_time)

def extract_ifstat_data_per_run(run_logs: str, utah_time: bool):
    start, end = get_test_start_end_time(os.path.join(run_logs, "client", "time.json"))
    return extract_ifstat_data_per_role(os.path.join(run_logs, "client"), start, end, utah_time), extract_ifstat_data_per_role(os.path.join(run_logs, "server"), start, end, utah_time)

def calculate_over_time_avg(runs_results):
    max_len = max(len(run) for run in runs_results)
    avg_cpu_usage = np.zeros(max_len)
    for i in range(max_len):
        count = 0
        for run_res in runs_results:
            if i < len(run_res):
                avg_cpu_usage[i] = avg_cpu_usage[i] + run_res[i]
                count += 1
        avg_cpu_usage[i] = avg_cpu_usage[i] / count
    return avg_cpu_usage

def extract_ifstat_per_testcase(testcase_logs: str, utah_time: bool):
    runs_dirs = [f for f in os.listdir(testcase_logs) if f.isdigit()]
    
    client_runs_per_secs = defaultdict(list)
    server_runs_per_secs = defaultdict(list)

    for run in runs_dirs:
        clients_runs, server_runs = extract_ifstat_data_per_run(os.path.join(testcase_logs, run), utah_time)
        for second, value in enumerate(clients_runs):
            client_runs_per_secs[second].append(value)
        for second, value in enumerate(server_runs):
            server_runs_per_secs[second].append(value)
    return  client_runs_per_secs, server_runs_per_secs

def axe_plot(ax: plt.Axes, cpu_util_per_sec: Dict[int, List[int]], title: str):
    data = []
    labels = list(cpu_util_per_sec.keys())
    labels.sort()
    for label in labels:
        data.append(cpu_util_per_sec[label])
    ax.boxplot(data, labels, showfliers=True, patch_artist=True, zorder=3)
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Bandwidth Usage (Mb/s)")
    ax.set_title(title)
    # Set x-ticks to every 10th value
    ax.set_xticks(np.arange(min(labels), max(labels)+1, 10))

    # Optionally, set x-tick labels to match the tick locations
    ax.set_xticklabels(np.arange(min(labels), max(labels)+1, 10))
    ax.grid(which='major', zorder=0, linestyle="dashed")

def plot_ifstat_impl(implementation_logs: str, impl_name: str, utah_time: bool):
    client_cpu_util, server_cpu_util = extract_ifstat_per_testcase(os.path.join(implementation_logs, testcase), utah_time)
    fig, axes = plt.subplots(2, 1, figsize=(12, 8), dpi=DPI, sharey=True)
    fig.subplots_adjust(hspace=0.5)  # Add margin between rows
    fig.suptitle(f"{impl_name}")
    axe_plot(axes[0], client_cpu_util, "Client")
    axe_plot(axes[1], server_cpu_util, "Server")
    plt.savefig(os.path.join(plots_directory, f"{impl_name}_{testcase}_{os.path.basename(expirements_logs).replace('-', '_')}_bandwidth_usage.pdf"), format="pdf", bbox_inches='tight')
    plt.savefig(os.path.join(plots_directory, f"{impl_name}_{testcase}_{os.path.basename(expirements_logs).replace('-', '_')}_bandwidth_usage.svg"), format="svg", bbox_inches='tight')
    #plt.yscale("log")
```

```python
if "quic_quic" in os.listdir(expirements_logs):    
    plot_ifstat_impl(os.path.join(expirements_logs, "quic_quic"), "QUIC", utah_time=False)
    plt.show()
```

```python
if "mpquic_mpquic" in os.listdir(expirements_logs):
    plot_ifstat_impl(os.path.join(expirements_logs, "mpquic_mpquic"), "MPQUIC", utah_time=False)
    plt.show()
```
